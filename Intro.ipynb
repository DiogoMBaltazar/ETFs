# Creating Indicators

```python
#----------------------------------------------------------
#                   FEATURES PARAMETERS
#----------------------------------------------------------
rsi_period = 21
macd_fast_period = 15
macd_slow_period = 26
macd_signal_period = 9
#----------------------------------------------------------

for etf in tqdm(etf_list):

  #ticker = yf.Ticker(etf)
  # total_assets = ticker.info['totalAssets']

  #etf_df = df_filtered[df_filtered['Symbol'] == etf]

  ### Exponential Moving Averages
  etf_final[f'EMA_9_{etf}'] = etf_final[f'Close_{etf}'].ewm(9).mean().shift()
  etf_final[f'EMA_30_{etf}'] = etf_final[f'Close_{etf}'].ewm(30).mean().shift()
  etf_final[f'EMA_50_{etf}'] = etf_final[f'Close_{etf}'].ewm(50).mean().shift()

  ### Simple Moving Averages
  etf_final[f'SMA_30_{etf}'] = etf_final[f'Close_{etf}'].rolling(30).mean().shift()
  etf_final[f'SMA_50_{etf}'] = etf_final[f'Close_{etf}'].rolling(50).mean().shift()
  etf_final[f'SMA_100_{etf}'] = etf_final[f'Close_{etf}'].rolling(100).mean().shift()
  etf_final[f'SMA_200_{etf}'] = etf_final[f'Close_{etf}'].rolling(200).mean().shift()

  ### Price changes // Returns 
  #### TODO // DOUBLE CHECK THIS
  etf_final[f'5D_Ret_{etf}'] = etf_final[f'Close_{etf}'].pct_change(periods = 5)
  etf_final[f'15D_Ret_{etf}'] = etf_final[f'Close_{etf}'].pct_change(periods = 15)
  etf_final[f'30D_Ret_{etf}'] = etf_final[f'Close_{etf}'].pct_change(periods = 30)
  etf_final[f'90D_Ret_{etf}'] = etf_final[f'Close_{etf}'].pct_change(periods = 90)
  etf_final[f'180D_Ret_{etf}'] = etf_final[f'Close_{etf}'].pct_change(periods = 180)
  etf_final[f'360D_Ret_{etf}'] = etf_final[f'Close_{etf}'].pct_change(periods = 360)

  # RSI
  etf_final[f'RSI_{etf}'] = ta.RSI(etf_final[f'Close_{etf}'], timeperiod=rsi_period)

  # Stochastic Oscillator
  #data["stochastic_k"], data["stochastic_d"] = talib.STOCH(data["High"], data["Low"], data["Close"], fastk_period=14, slowk_period=3, slowd_period=3)

  # ATX

  # Flow = # SHARES OUTSTANDING
  etf_volume = etf_final[f'Volume_{etf}']
  etf_close = etf_final[f'Close_{etf}']

  # Compute the daily flows for each ETF
  etf_final[f'Daily_Flows_{etf}'] = etf_volume * etf_close.diff()

  # Dont forget to convert to log returns 
  etf_final[f'Log_15D_Ret_{etf}'] = np.log(1 + etf_final[f'15D_Ret_{etf}'])
  etf_final[f'Log_5D_Ret_{etf}'] = np.log(1 + etf_final[f'5D_Ret_{etf}'])
  etf_final[f'Log_30D_Ret_{etf}'] = np.log(1 + etf_final[f'30D_Ret_{etf}'])
  etf_final[f'Log_90D_Ret_{etf}'] = np.log(1 + etf_final[f'90D_Ret_{etf}'])
  etf_final[f'Log_180D_Ret_{etf}'] = np.log(1 + etf_final[f'180D_Ret_{etf}'])
  etf_final[f'Log_360D_Ret_{etf}'] = np.log(1 + etf_final[f'360D_Ret_{etf}'])
  #----------------------------------------------------------
  #----------------------------------------------------------


  print(etf_final.tail(2))
```
